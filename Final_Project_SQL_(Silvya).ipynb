## **SQL statement untuk menemukan jumlah customer per state**

Menemukan jumlah customer per country menggunakan SQL Command COUNT dan GROUP BY

```python
sql6 = """SELECT customer_state, COUNT(customer_id) as total_customer
      FROM public.oilist_customer GROUP BY customer_state;"""
customer_per_state = pd.read_sql_query(sql6, conn)
customer_per_state
```

## **SQL statement untuk menemukan jumlah total order per state**

Menemukan jumlah total order per state menggunakan SQL command JOIN, COUNT dan GROUP BY

```python
sql7 = """SELECT t2.customer_state, COUNT(t1.order_id)
      FROM public.oilist_orders t1
      INNER JOIN public.oilist_customer t2
      ON t1.customer_id = t2.customer_id
      GROUP BY t2.customer_state
      ORDER BY customer_state asc;"""
order_per_state = pd.read_sql_query(sql7, conn)
order_per_state
```

Menggunakan **COUNT** dari order_id dari tabel_orders di **JOIN** ke tabel table_order_detail untuk seller_id. **GROUP BY** berdasarkan Month. Untuk melihat perform seller setiap bulan.

```python
# sql8 = """SELECT t1.seller_id,
#         CAST(EXTRACT(MONTH FROM TO_DATE(t1.order_purchase_timestamp, 'DD/MM/YYYY')) AS INT) as order_month,
#        COUNT(t3.order_id) as total_order
#        FROM public.oilist_sellers t1
#        INNER JOIN public.oilist_orders_detail ON t1.seller_id = t2.seller_id
#        INNER JOIN public.oilist_orders t3 ON t2.order_id = t3.order_id
#        GROUP BY t1.seller_id, order_month
#        ORDER BY seller_id, order_month;"""
sql8 = """SELECT t1.seller_id, CAST(EXTRACT(MONTH FROM TO_DATE(t3.order_purchase_timestamp, 'DD/MM/YYYY')) AS INT) as order_month, 
         COUNT(t3.order_id) as total_order
         FROM public.oilist_sellers t1
         INNER JOIN public.oilist_orders_detail t2
         ON t1.seller_id = t2.seller_id 
         INNER JOIN public.oilist_orders t3
         ON t2.order_id = t3.order_id
         GROUP BY t1.seller_id,  order_month
         ORDER BY seller_id, order_month;"""
order_per_month = pd.read_sql_query(sql8,conn)
order_per_month
```

## **SQL statement untuk menemukan TOP 10 customer by total_price dari setiap state**

Menggunakan **COUNT** dari total_price lalu **JOIN** untuk table order dan user by order_id **GROUP BY** country, **ORDER BY DESC** lalu di limit sebanyak 10.

```python
sql9 = """SELECT t1.customer_id, SUM(t3.price) as total_price,t1.customer_state
          FROM public.oilist_customer t1
          INNER JOIN public.oilist_orders t2
          ON t1.customer_id = t2.customer_id
          INNER JOIN public.oilist_orders_detail t3
          ON t2.order_id = t3.order_id
          GROUP BY t1.customer_id,t1.customer_state
          ORDER BY total_price DESC LIMIT 10;"""
top_10_buyer = pd.read_sql_query(sql9,conn)
top_10_buyer
```

## **SQL statement untuk menemukan latest order_purchase_timestamp dari setiap customer**

Menggunakan table order yang di **GROUP BY** order_time dan customer_id lalu di order by **ORDER BY DESC** limit 1

```python
sql10 = """SELECT customer_id, MAX(TO_TIMESTAMP(order_purchase_timestamp, 'DD/MM/YYYY HH24.MI')) as order_purchase_timestamp
        FROM public.oilist_orders 
        GROUP BY customer_id
        ORDER BY order_purchase_timestamp DESC;"""
latest_order_buyer = pd.read_sql_query(sql10,conn)
latest_order_buyer
```

## **SQL statement untuk menemukan jumlah masing-masing order_status dari masing-masing customer**

Menggunakan kolom order_status untuk menentukan jumlah masing-masing order status dari masing-masing customer

```python
sql11 = """SELECT customer_id,order_status,COUNT(order_status) as jumlah
        FROM public.oilist_orders 
        GROUP BY customer_id,order_status
        ORDER BY customer_id ASC;"""
order_per_customer = pd.read_sql_query(sql11,conn)
order_per_customer
```
